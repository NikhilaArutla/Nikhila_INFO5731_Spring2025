```python
import pandas as pd
import numpy as np
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# ---------------------- Step 1: Load and Clean Dataset ---------------------- #
print(" Loading and cleaning dataset...")

df = pd.read_csv('ANNOTATED_SENTIMENT_DATA_A3_Q4.csv')
df = df.dropna(subset=['clean_text'])
df = df.drop_duplicates(subset=['clean_text'])
df['clean_text'] = df['clean_text'].str.replace("no review added viewer", "", regex=False).str.strip()
df = df[df['clean_text'].str.strip() != '']

# Print dataset size
print(f" Final dataset size after cleaning: {df.shape[0]} rows\n")

# Convert to list
docs = df['clean_text'].tolist()

# ---------------------- Step 2: Initial BERTopic with K=10 ---------------------- #
print(" Training BERTopic model with K=10...")

initial_model = BERTopic(nr_topics=10, verbose=True)
initial_topics, _ = initial_model.fit_transform(docs)

# Display initial topics
print("\n Initial Topics (K=10):")
for topic in set(initial_topics):
    if topic != -1:
        words = initial_model.get_topic(topic)
        word_str = ", ".join([w for w, _ in words])
        print(f"\nTopic {topic}: {word_str}")

# Safe visualizations
if len(initial_model.get_topics()) > 1 and len(set(initial_topics)) > 1:
    try:
        initial_model.visualize_topics().show()
        initial_model.visualize_barchart().show()
    except ValueError as e:
        print(f" Error visualizing topics: {e}")
else:
    print(" Not enough distinct topics to visualize.")

# ---------------------- Step 3: Find Optimal K via Coherence ---------------------- #
def calculate_coherence_score(docs, start_k=5, end_k=15):
    print("\n Calculating coherence scores to find optimal K...")
    coherence_scores = []
    for k in range(start_k, end_k + 1):
        print(f"   Evaluating K={k}")
        temp_model = BERTopic(nr_topics=k, verbose=False)
        temp_topics, _ = temp_model.fit_transform(docs)

        topics_tokens = [
            [word for word, _ in temp_model.get_topic(topic_id)]
            for topic_id in set(temp_topics) if topic_id != -1 and temp_model.get_topic(topic_id)
        ]
        if not topics_tokens:
            continue

        analyzer = temp_model.vectorizer_model.build_analyzer()
        tokenized_docs = [analyzer(doc) for doc in docs]

        dictionary = Dictionary(tokenized_docs)
        corpus = [dictionary.doc2bow(text) for text in tokenized_docs]

        cm = CoherenceModel(topics=topics_tokens, texts=tokenized_docs, corpus=corpus,
                            dictionary=dictionary, coherence='u_mass')
        coherence = cm.get_coherence()
        coherence_scores.append((k, coherence))
        print(f"     Coherence score: {coherence:.4f}")
    return coherence_scores

coherence_scores = calculate_coherence_score(docs)

# Plot coherence scores
if coherence_scores:
    ks, scores = zip(*coherence_scores)
    plt.figure(figsize=(8, 4))
    plt.plot(ks, scores, marker='o', color='teal')
    plt.title("Coherence Score vs Number of Topics (K)")
    plt.xlabel("Number of Topics (K)")
    plt.ylabel("U_Mass Coherence Score")
    plt.grid(True)
    plt.show()

    # ---------------------- Step 4: Retrain with Optimal K ---------------------- #
    optimal_k = max(coherence_scores, key=lambda x: x[1])[0]
    print(f"\n Optimal number of topics based on coherence: K={optimal_k}")

    print("\n Retraining BERTopic with optimal K...")
    topic_model = BERTopic(nr_topics=optimal_k, verbose=True)
    topics, _ = topic_model.fit_transform(docs)

    print("\n Final Interpreted Topics:")
    for topic in set(topics):
        if topic != -1:
            words = topic_model.get_topic(topic)
            word_str = ", ".join([w for w, _ in words])
            print(f"\nTopic {topic}: {word_str}")

            # Word Cloud
            wordcloud = WordCloud(width=800, height=400, background_color="white").generate(word_str)
            plt.figure(figsize=(8, 4))
            plt.imshow(wordcloud, interpolation="bilinear")
            plt.axis("off")
            plt.title(f"Word Cloud for Topic {topic}")
            plt.show()

    # Visualizations
    print("\n Generating topic visualizations...")
    if len(set(topics)) > 1:
        topic_model.visualize_topics().show()
        topic_model.visualize_barchart().show()
        if optimal_k > 2:
            topic_model.visualize_hierarchy().show()
        else:
            print(" Not enough topics for hierarchy visualization.")
    else:
        print(" Not enough distinct topics to generate visualizations.")
else:
    print(" No valid coherence scores found. Cannot determine optimal K.")
```

```python
from sklearn.datasets import fetch_20newsgroups

# Load 2000 random documents
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
docs = newsgroups.data[:2000]

# Preprocessing function
def preprocess(text):
    text = re.sub(r'\W+', ' ', text.lower())  # Remove special chars + lowercase
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in stopwords.words('english') and len(t) > 3]  # Remove short words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens

processed_docs = [preprocess(doc) for doc in docs]
```

```python
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
def train_evaluate_lda(corpus, dictionary, texts, start=5, limit=20, step=1):
    models = []
    coherences = []

    for num_topics in range(start, limit + 1, step):
        model = LdaModel(
            corpus=corpus,
            id2word=dictionary,
            num_topics=num_topics,
            random_state=42,
            passes=10
        )
        coherence_model = CoherenceModel(
            model=model,
            texts=texts,
            dictionary=dictionary,
            coherence='c_v'
        )
        coherence_score = coherence_model.get_coherence()
        models.append(model)
        coherences.append(coherence_score)
        print(f"Topics: {num_topics}, Coherence Score: {coherence_score:.4f}")

    return models, coherences

models, scores = train_evaluate_lda(corpus, dictionary, processed_docs)
best_index = np.argmax(scores)
optimal_topics = range(5, 21)[best_index]
print(f"\nOptimal Topics: {optimal_topics}, Best Coherence: {scores[best_index]:.4f}")
```
